```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegressionCV
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
df = pd.read_csv("C:/Datascience/train_mobile.csv")
df
```

```python
df.columns
```

```python
df['price_range'].unique()
```

Means that the column 'price_range' in the DataFrame contains four distinct categories:

â€¢ 0 â†’ Low price range
â€¢ 1 â†’ Medium price range
â€¢ 2 â†’ High price range
â€¢ 3 â†’ Very high price range

These values are likely labels for classification, and  machine learning model
is trained to predict one of these four classes based on mobile specifications.

```python
x = df[['battery_power', 'blue', 'clock_speed', 'dual_sim', 'fc', 'four_g',
       'int_memory', 'm_dep', 'mobile_wt', 'n_cores', 'pc', 'px_height',
       'px_width', 'ram', 'sc_h', 'sc_w', 'talk_time', 'three_g',
       'touch_screen', 'wifi']]

y = df['price_range']
```

```python
# train/test split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

ðŸ”¹ What it means:
â€¢ 80% of data goes to training (X_train, y_train)
â€¢ 20% goes to testing (X_test, y_test)
random_state=42 ensures the split is reproducible every time you run it

```python
knn_model = KNeighborsClassifier(n_neighbors=5)
```

```python
knn_model.fit(x_train,y_train)
```

ðŸ”¹ What it means:
â€¢ Initializes a KNN classifier with 5 neighbors
â€¢ Trains it on your training data (X_train, y_train)
â€¢The model will classify test samples based on the 5 closest training points

```python
y_pred = knn_model.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
```

ðŸ”¹ What it means:
â€¢ Shows precision, recall, f1-score, and support for each class (0 to 3)
â€¢ The model achieved 94% accuracy
â€¢ Performance is strong across all price categories

```python
print(confusion_matrix(y_test,y_pred))
```

ðŸ”¹ What it means:

â€¢ Rows = actual classes, columns = predicted classes
â€¢ Diagonal values (103, 85, 84, 105) = correct predictions
â€¢ Off-diagonal values = misclassifications
â€¢ The model is performing well with minimal errors

```python
print(accuracy_score(y_test,y_pred)*100)
```
